```python
#2.4 Podemos realizar la misma operacin con la funcin *scale* del paquete preprocessing
def preprocesamiento_scale(df):
    import pandas as pd
    from sklearn.preprocessing import scale
    df_arr =  scale(df)
    df =  pd.DataFrame(df_arr, index = df.index, columns=df.columns)
    
    return df, df.mean()

```

```python
# 3. Generar caractersticas que se consideren interesantes para reconocer el patron a partir de los datos
# de la ventana temporal (ver comentarios mas adelante).
# Sobre la generacion de caractersticas se recomienda generar funciones sencillas que calculen valores
# que ayuden a distinguir unas series de otras, independientemente del patron que se este buscando. Como
# los ventanas tendran distinta escala de precios, es necesario que las caractersticas esten normalizadas.


# 3.1 Voy leyendo pkl por pkl y le paso las caracteristicas que he creado para poner valor a la caracteritica pro cada grafica.
# Despues salvo toda la informacion de las catateritticas y las etiquetas en un df que lo guardo.
columnas = ['Ejemplo_n','Primer_dia','Ultimo_dia','Valor_Etiqueta','condicion_1', 'condicion_2','condicion_3_1',
            'condicion_3_2','condicion_4','condicion_4_1','condicion_5','condicion_5_1']
df = pd.DataFrame(0, index=np.arange(n), columns = columnas)
ibex['lista'] =range(0,ibex.shape[0])
for i in range(0,ejemplos.shape[0]):
    # extraigo pkl
    with open(f'OHLC_pck_Ej_5_1/{i}_OHLC.pkl', 'rb') as handle:
        bm = pickle.load(handle)

    


# Tomas recomendo estandarizar los datos OHLC para la condicion 3 que el propuso. Ver punto 2.4
# Lo hago ahora para todos los ohlc que me van saliendo
    a,b = preprocesamiento_scale(bm.iloc[:,0:4])
    bm.close = a.close
    bm.high = a.high
    bm.low = a.low
    bm.open = a.open


# Algunos ejemplos pueden ser:

#  CONDICION_1: Numero de das que pasan entre el maximo y el mnimo de la ventana
# Toma el primer maximo o el primer minimo ya que aveces hay dos veces el maximo y el minimo

    MAX = bm.index[bm.high == bm.high.max()][0]
    MIN = bm.index[bm.low == bm.low.min()][0]
    condicion_1 =abs(int(bm[bm.index == MAX].lista.values) - int(bm[bm.index == MIN].lista.values))
       
    
#  CONDICION_2: Si ocurre primero el maximo o el mnimo
    posicion_max = bm.index[bm.high == bm.high.max()][0] - bm.index[0]
    posicion_min = bm.index[bm.low == bm.low.min()] - bm.index[0]
    if (posicion_max > posicion_min):
        condicion_2 = 1 #primero es en minimo
    else: 
        condicion_2 = 0 #primero es en maximo o el mismo dia
    condicion_2
    
    
#  CONDICION_3: Numero de veces en la ventana que se llega a cierto umbral ( lo he puesto que lo supera; no cruzar)
    umbral = bm.open[0] * 0.05 # umbral a un 5% del valor de apertura
    umbral_superior = bm.open[0] + umbral
    umbral_inferior =  bm.open[0] - umbral
    numero_de_veces_superar_maximo = 0
    numero_de_veces_superar_minimo = 0
    for i_2 in range(0,bm.shape[0]):
        if bm.high[i_2] > umbral_superior:
            numero_de_veces_superar_maximo = numero_de_veces_superar_maximo + 1
        if bm.low[i_2] < umbral_inferior:
            numero_de_veces_superar_minimo  = numero_de_veces_superar_minimo + 1
    condicion_3_1 = numero_de_veces_superar_maximo
    condicion_3_2 = numero_de_veces_superar_minimo
    
    #  Si el volumen en el suelo ( cuando da minimo) es mayor a un umbral respecto al volumen medio de del periodo. 
    # Lo suyo seria hacerlo con el minimo y el segundo minimo. Si en los dos el volumen es mayor al volumen medio 
    # de del periodo mas un umbral
    
#  CONDICIONES_4: Si el volumen en el suelo ( cuando da minimo) es mayor a la media de todos los volumenes de ese periodo,
# considero que esta tiene doble suelo. Pero este minimo tiene que estar en la primeros 3/4 de la grafica
    condicion_4 = 0
    condicion_4_1 = 0
    if (bm.loc[bm.low == bm.low.min()].index - bm.index[0]).days.tolist()[0] < (window * 0.75): 
        if bm.loc[bm.low == bm.low.min()]['vol'][0] - np.mean(bm['vol']) > 0:
            # print(bm.loc[bm.low == bm.low.min()]['vol'][0] - np.mean(bm['vol']))
            # print(np.mean(bm['vol']))
            # print((bm.loc[bm.low == bm.low.min()]['vol'][0] - np.mean(bm['vol']))/(np.mean(bm['vol'])))
            condicion_4 = 1
        condicion_4_1 = (bm.loc[bm.low == bm.low.min()]['vol'][0] - np.mean(bm['vol']))/(np.mean(bm['vol']))

#  CONDICIONES_5: Si el volumen en el segundo suelo ( cuando da minimo) es mayor a la media de todos los volumenes de ese periodo,
# considero que esta tiene doble suelo. Pero este segundo minimo tiene que estar en la primeros 85% de la grafica
# No me ha dado tiempo a desaroyarlo del todo que deberia tener que el segundo suelo estaria en una distancia de unos 
# mas de 5 dias de cotizacion para considerarlo otro suelo
    lows = list(bm.low)
    lows.sort(reverse=False)
    condicion_5 = 0
    condicion_5_1 = 0
    if (bm.loc[bm.low == lows[1]].index - bm.index[0]).days.tolist()[0] < (window * 0.85): 
        if bm.loc[bm.low == lows[1]]['vol'][0] - np.mean(bm['vol']) > 0:
            condicion_5 = 1
        condicion_5_1 = (bm.loc[bm.low == lows[1]]['vol'][0] - np.mean(bm['vol']))/(np.mean(bm['vol']))

        
# Meto todo en un df
    columnas = ['Ejemplo_n','Valor_Etiqueta','condicion_1', 'condicion_2','condicion_3_1',
               'condicion_3_2','condicion_4','condicion_4_1','condicion_5','condicion_5_1']
    df.loc[i,'condicion_1'] = condicion_1
    df.loc[i,'condicion_2'] = condicion_2
    df.loc[i,'condicion_3_1'] = condicion_3_1
    df.loc[i,'condicion_3_2'] = condicion_3_2
    df.loc[i,'condicion_4'] = condicion_4
    df.loc[i,'condicion_4_1'] = condicion_4_1
    df.loc[i,'condicion_5'] = condicion_5
    df.loc[i,'condicion_5_1'] = condicion_5_1                                                                                         
    df.loc[i,'Valor_Etiqueta'] = bm['Valor_Etiqueta'][0]
    df.loc[i,'Ejemplo_n'] = i    
    df.loc[i,'Primer_dia'] = bm.index[0].strftime('%Y-%m-%d')
    df.loc[i,'Ultimo_dia'] = bm.index[-1].strftime('%Y-%m-%d')
    ve= bm['Valor_Etiqueta'][0]
    pri_dia = bm.index[0].strftime('%Y-%m-%d')
    ult_dia = bm.index[-1].strftime('%Y-%m-%d')
df
```

```python
# Elijo el rboles de decisin
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree

# 4.1 establezco que es features y target
features = df_maestro.loc[:,df_maestro.columns[5:13]] #es un df
target = df_maestro.loc[:,df_maestro.columns[4]] #es una serie

# 4.2 hago un scale de este df pq hay condiciones con mucha mas rangeabilidad que otras condiciones
# me da mejores valores de accuracy, aucroc, roc, matriz de confusion....

##############################
features_2 = preprocesamiento_StandardScaler(features.iloc[:,[0,2,3,5,7]])

features.condicion_1 = features_2.condicion_1
features.condicion_3_1 = features_2.condicion_3_1
features.condicion_3_2 = features_2.condicion_3_2
features.condicion_4_1 = features_2.condicion_4_1
features.condicion_5_1 = features_2.condicion_5_1
################

#hago el split a un 40%
x_train, x_test, y_train, y_test = train_test_split(features.values,
                                                    target.values,
                                                    test_size=0.4,
                                                    stratify=target.values)
```

```python
# 5. Reportar los resultados de evaluacion que pueden incluirse en un notebook con el codigo para
# generarlos. La evaluacion debe incluir:
# (a) Una matriz de confusion
# (b) Resultado de accuracy y AUROC
# (c) Una curva ROC

# 5.1 Entreno el algoritmo Tree
test_predicted = dTree.predict(x_test)
etiquetas_VS_prediccion = pd.DataFrame({'label': y_test, 'predicted': test_predicted})

# 5.2 Calculo la matriz de confusion
from sklearn.metrics import confusion_matrix, accuracy_score
matriz_de_confusion_1 = confusion_matrix(y_test, test_predicted)
print(matriz_de_confusion_1)

def matriz_confusion(y_real, y_pred):
    real = pd.Series(y_real, name='Real')
    predicted = pd.Series(y_pred, name="Predicted")
    return pd.crosstab(predicted, real)

matriz_confusion(y_test, test_predicted)
```

```python
# 5.3 Calculo el Resultado de accuracy y AUROC

# 5.3.1 Accuracy
print(f'Accuracy es: {accuracy_score(y_test, test_predicted)}')
```

```python
# 5.3.2 Accuracy y AUROC. Otra forma.
import sklearn
from sklearn.model_selection import cross_validate, KFold
from sklearn.metrics import accuracy_score, roc_auc_score
scoring = ['accuracy', 'roc_auc']
scores = cross_validate(dTree, features, target, scoring=scoring, cv=4)
pd.DataFrame(scores), pd.DataFrame(scores).mean()
```

```python
# 5.3.3 otra forma de calcular esas metricas con Kfold
kfold = KFold(n_splits=4, shuffle=True)
acc = []
auc = []
features_arr = features.to_numpy()
for train, test in kfold.split(features_arr):
    
    fold_train_x = features_arr[train]
    fold_train_y = target[train]
    dTree.fit(fold_train_x, fold_train_y)
    
    fold_test_x = features_arr[test]
    fold_test_y = target[test]
    
    i_acc = dTree.score(fold_test_x, fold_test_y)
    
    i_class = np.where(dTree.classes_ == 'yes')[0][0]
    print('class:', i_class)
    fold_test_pred = dTree.predict_proba(fold_test_x)[:,i_class]
    i_auc = roc_auc_score(fold_test_y, fold_test_pred)

    auc.append(i_auc)
    acc.append(i_acc)
result = pd.DataFrame({'auc':auc, 
                       'accuracy':acc})
result, result.mean() # (b) Resultado de accuracy y AUROC
```

```python
# 5.4 Curva ROC

from sklearn.metrics import roc_curve, auc
test_pred = dTree.predict_proba(x_test)
test_posprob = test_pred[:,1]
train_pred = dTree.predict_proba(x_train)
train_posprob = train_pred[:,1]
train_fpr, train_tpr, train_thr = roc_curve(y_train, train_posprob, pos_label='yes')
test_fpr, test_tpr, test_thr = roc_curve(y_test, test_posprob, pos_label='yes')
fig, ax = plt.subplots(figsize=(5,5))
plt.plot(train_fpr, train_tpr, color='r', label='train')
plt.plot(test_fpr, test_tpr, color='g', label='test')
plt.plot([0,1],[0,1])
plt.xlim([0.0, 1.0])
plt.ylim([0.0,1.0])
plt.legend(loc='lower right')
```
